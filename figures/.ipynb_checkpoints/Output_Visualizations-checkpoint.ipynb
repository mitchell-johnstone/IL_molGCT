# Visualization of Model Output
The GCT produces chemicals, we need to visualize the results.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#!pip install rdkit
from rdkit import Chem
# import rdkit.Chem.Draw
```

### Get Output data

```python
output_files = {
    'atom: 30000':'atom_IL_moses_bench2_lat=128_epo=50_k=4_20230723.csv',
    'atom: 10 cond':'atom_moses_bench2_10conds_lat=128_epo=50_k=4_20230722.csv',
    'smilespe: 30000':'smilespe_IL_moses_bench2_lat=128_epo=50_k=4_20230723.csv',
    'smilespe: 10 cond':'smilespe_moses_bench2_10conds_lat=128_epo=50_k=4_20230722.csv'
}
output_files
```

```python
def getDetails(name, file):
    out = pd.read_csv("../outputs/"+file)
    print("\n---------------------------------------------")
    print(name,"\n")
    print(out.info())
    print(out.head())
    print("Number of Valid chemicals:", out.Valid.sum())
    print("Validity: ", out.Valid.sum() / len(out.Valid))
    return out[out.Valid==1]

output_dfs = dict(zip(output_files.keys(), (getDetails(name, output_files[name]) for name in output_files.keys())))
# getDetails('atom: 30000', output_files['atom: 30000'])
# getDetails('smilespe: 30000', output_files['smilespe: 30000'])
output_dfs.keys()
```

### Get original input data

```python
inp = pd.read_csv("../data/IL/train.csv")
print(inp.info())
inp.head()
```

### Check the 3D distribution of datapoints

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(inp.DynViscosity, inp.Density, inp.ElecConductivity)
```

### Make all smiles canon

```python
for name, out in output_dfs.items():
    out["Canon"] = out.SMILES.apply(Chem.CanonSmiles)
    print(len(set(out.Canon)))
```

```python
# in_SMILES = []
# for smiles in inp.src:
#     try:
#         in_SMILES.append(Chem.CanonSmiles(smiles))
#     except:
#         in_SMILES.append("")
#         pass
# inp["Canon"] = in_SMILES
inp["Canon"] = inp["src"].apply(Chem.CanonSmiles)
print(len(set(inp.Canon)))
```

### Filter out known chemicals

```python
output_unique = {}
for name, out in output_dfs.items():
    novel = set(out.Canon).difference(set(inp.Canon))
    mask = out.Canon.apply(lambda s: s in novel)
    output_unique[name] = out[mask]
    print(name)
    print("Number novel: ",len(novel))
    print("Novelty: ", len(novel)/len(set(out.Canon)))
    print("First 5 ILs...")
    for n in list(novel)[:10]:
        print("\t", n)
    print()
```

```python
for name, unique in output_unique.items():
    print("\n---------------------------------------------")
    print(name,"\n")
    print(unique.info())
    print(unique.head())
```
