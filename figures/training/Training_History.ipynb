# Visualization of Training History

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
## Set constants
fract_val = 0.15
figurewidth, figureheight = 16, 8
font = {'family': 'sans-serif', 'weight': 'light', 'size': 14}
tickfont = {'family': 'sans-serif', 'weight': 'light', 'size': 12}
plot_box_thickness = 2
figuredpi = 300
```

```python
for d in os.listdir('../../weights'):
    if os.path.isdir('../../weights/'+d) and ("smiles" in d or "atom" in d):
        folder = "../../weights/"+d+"/train_history/"
        print(folder)
        for file in os.listdir(folder):
            try:
                df = pd.read_csv(folder+file)
            except:
                continue
            plt.figure()
            plt.plot(df["epoch"], df["total_loss"], label="Training loss")
            plt.plot(df["epoch"], df["total_loss_te"], label="Validation loss")
            plt.xlabel("Epoch", fontdict=font)
            plt.ylabel("Loss", fontdict=font)
            plt.title("Training History", fontdict=font)
            plt.legend()
            plt.show()
            plt.savefig(folder+"Training_History.png", dpi=figuredpi, bbox_inches='tight')
```
